```python
import pandas as pd
import pandas.io.sql as sqlio
import psycopg2

# Retrieving data from postgres for the year 2017
#sql = '''select county,term,sum(total_attending_suny_institutions) as "tot_stu_county" from edu_ny_data group by term,county order by county,term;'''
sql = '''select enyd.county,
sum(enyd.total_attending_suny_institutions) as "tot_stu",
sum(nym.crrcgc) as "recipients",
sum(nycr.total_crimes) as "tot_crime"
from edu_ny_data as "enyd" join newyorkmentalhealth as "nym" 
on lower(trim(enyd.county)) = lower(trim(nym.county_label))
join nycrime as nycr on lower(trim(enyd.county)) = lower(trim(nycr.county))
where enyd.term = '2017'
and nym.service_year = '2017'
and nycr.year='2017'
group by enyd.county order by enyd.county;'''

sql_vehicle = '''select county,count(1) as "vehicle_cnt" from rebate where to_date(submitted_date,'yyyy-mm-dd') 
between to_date('2017-01-01','yyyy-mm-dd')
and to_date('2017-12-31','yyyy-mm-dd') group by county order by county;'''
try:
    dbConnection = psycopg2.connect(
        user = "postgres",
        password = "lightyear",
        host = "localhost",
        port = "5432",
        database = "dapgroupm")
    df = sqlio.read_sql_query(sql, dbConnection)
    df1 = sqlio.read_sql_query(sql_vehicle, dbConnection)
except (Exception, psycopg2.Error) as dbError :
    print ("Error:", dbError)
finally:
    if(dbConnection): 
        dbConnection.close()
        
df['stu_pct'] = (df.tot_stu/df.tot_stu.sum())*100
df['recipient_pct'] = (df.recipients/df.recipients.sum())*100
df['crime_pct'] = (df.tot_crime/df.tot_crime.sum())*100
df['tot_vehicle_pct'] = (df1.vehicle_cnt/df1.vehicle_cnt.sum())*100
print(df.shape)

print(df1.shape)
```

```python
# Retrieving data from postgres for the year 2018

sql = '''select enyd.county,
sum(enyd.total_attending_suny_institutions) as "tot_stu",
sum(nym.crrcgc) as "recipients",
sum(nycr.total_crimes) as "tot_crime"
from edu_ny_data as "enyd" join newyorkmentalhealth as "nym" 
on lower(trim(enyd.county)) = lower(trim(nym.county_label))
join nycrime as nycr on lower(trim(enyd.county)) = lower(trim(nycr.county))
where enyd.term = '2018'
and nym.service_year = '2018'
and nycr.year='2018'
group by enyd.county order by enyd.county;'''

sql_vehicle = '''select county,count(1) as "vehicle_cnt" from rebate where to_date(submitted_date,'yyyy-mm-dd') 
between to_date('2017-01-01','yyyy-mm-dd')
and to_date('2017-12-31','yyyy-mm-dd') group by county order by county;'''
try:
    dbConnection = psycopg2.connect(
        user = "postgres",
        password = "lightyear",
        host = "localhost",
        port = "5432",
        database = "dapgroupm")
    df_18 = sqlio.read_sql_query(sql, dbConnection)
    df1_18 = sqlio.read_sql_query(sql_vehicle, dbConnection)
except (Exception, psycopg2.Error) as dbError :
    print ("Error:", dbError)
finally:
    if(dbConnection): 
        dbConnection.close()
        
df_18['stu_pct'] = (df_18.tot_stu/df_18.tot_stu.sum())*100
df_18['recipient_pct'] = (df_18.recipients/df_18.recipients.sum())*100
df_18['crime_pct'] = (df_18.tot_crime/df_18.tot_crime.sum())*100
df_18['tot_vehicle_pct'] = (df1_18.vehicle_cnt/df1_18.vehicle_cnt.sum())*100
print(df_18.shape)

print(df1_18.shape)
```

```python
# Exploring data for visualization in individual data
# Top 10 counties for student enrollment

import pandas as pd
import pandas.io.sql as sqlio
import psycopg2

# Retrieving data from postgres for the year 2017
#sql = '''select county,term,sum(total_attending_suny_institutions) as "tot_stu_county" from edu_ny_data group by term,county order by county,term;'''
sql_stu_17 = '''select county,term,sum(total_attending_suny_institutions) as "tot_student_enrollment" 
from edu_ny_data where term = '2017' group by term,county 
order by tot_student_enrollment desc fetch first 10 rows only;'''
# Retrieving data from postgres for the year 2018
sql_stu_18 = '''select county,term,sum(total_attending_suny_institutions) as "tot_student_enrollment" 
from edu_ny_data where term = '2018' group by term,county 
order by tot_student_enrollment desc fetch first 10 rows only;'''

# Retrieving data from postgres for the year 2017
sql_crime_17 = '''select county,year,sum(total_crimes) as "tot_crimes" 
from nycrime where year = '2017' group by county,year order by tot_crimes desc fetch first 10 rows only;'''
# Retrieving data from postgres for the year 2018
sql_crime_18 = '''select county,year,sum(total_crimes) as "tot_crimes" 
from nycrime where year = '2018' group by county,year order by tot_crimes desc fetch first 10 rows only;'''

# Retrieving data from postgres for the year 2017
sql_vehicle_17 = '''select county,'2017' as "year",count(1) as "tot_vehicles" from rebate where to_date(submitted_date,'yyyy-mm-dd') 
between to_date('2017-01-01','yyyy-mm-dd')
and to_date('2017-12-31','yyyy-mm-dd') group by county order by tot_vehicles desc fetch first 10 rows only;'''
# Retrieving data from postgres for the year 2018
sql_vehicle_18 = '''select county,'2018' as "year",count(1) as "tot_vehicles" from rebate where to_date(submitted_date,'yyyy-mm-dd') 
between to_date('2018-01-01','yyyy-mm-dd')
and to_date('2018-12-31','yyyy-mm-dd') group by county order by tot_vehicles desc fetch first 10 rows only;'''

# Retrieving data from postgres for the year 2017
sql_health_17 = '''select county_label as "county",service_year,sum(crrcgc) as "recipients" 
from newyorkmentalhealth where service_year='2017'
group by county_label,service_year order by recipients desc fetch first 10 rows only;'''
# Retrieving data from postgres for the year 2018
sql_health_18 = '''select county_label as "county",service_year,sum(crrcgc) as "recipients" 
from newyorkmentalhealth where service_year='2018'
group by county_label,service_year order by recipients desc fetch first 10 rows only;'''


try:
    dbConnection = psycopg2.connect(
        user = "postgres",
        password = "lightyear",
        host = "localhost",
        port = "5432",
        database = "dapgroupm")
    df_stu_17 = sqlio.read_sql_query(sql_stu_17, dbConnection)
    df_stu_18 = sqlio.read_sql_query(sql_stu_18, dbConnection)
    
    df_crime_17 = sqlio.read_sql_query(sql_crime_17, dbConnection)
    df_crime_18 = sqlio.read_sql_query(sql_crime_18, dbConnection)
    
    df_vehicle_17 = sqlio.read_sql_query(sql_vehicle_17, dbConnection)
    df_vehicle_18 = sqlio.read_sql_query(sql_vehicle_18, dbConnection)
    
    df_health_17 = sqlio.read_sql_query(sql_health_17, dbConnection)
    df_health_18 = sqlio.read_sql_query(sql_health_18, dbConnection)
except (Exception, psycopg2.Error) as dbError :
    print ("Error:", dbError)
finally:
    if(dbConnection): 
        dbConnection.close()

```
